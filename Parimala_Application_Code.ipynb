```python
def Statisticsdata():
    data=pd.read_csv("WisCon_Stat.csv")
    data=data.drop('Unnamed: 0',axis=1)
    fet,infl,chv=data['Feature'].tolist(),data['Influence'].tolist(),data['Chi2'].tolist()
    for x in range(len(infl)):
        infl[x]=round(infl[x],2)*100
    return fet,infl,chv
```
